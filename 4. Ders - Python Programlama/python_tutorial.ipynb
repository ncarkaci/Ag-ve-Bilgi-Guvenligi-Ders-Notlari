Lists are mutable arrays. Let's see how they work.

List slicing is a useful way to access a slice of elements in a list.

```python
numbers = [0, 1, 2, 3, 4, 5, 6]

# Slices from start index (inclusive) to end index (exclusive)
print(numbers[0:3])
```

```python
# Negative index wraps around, start counting from the end of list
print(numbers[-1])
print(numbers[-3:])
print(numbers[3:-2])
```

Next, let's look at some matrix operations. Let's take an element-wise product (Hadamard product).

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 3], [3, 3]])
print(A)
print(B)
print("---")
print(A * B)
```

We can do matrix multiplication with np.matmul or @.

```python
# One way to do matrix multiplication
print(np.matmul(A, B))

# Another way to do matrix multiplication
print(A @ B)
```

We can take the dot product or a matrix vector product with np.dot.

```python
u = np.array([1, 2, 3])
v = np.array([1, 10, 100])

print(np.dot(u, v))

# Can also call numpy operations on the numpy array, useful for chaining together multiple operations
print(u.dot(v))
```

```python
W = np.array([[1, 2], [3, 4], [5, 6]])
print(v.shape)
print(W.shape)

# This works.
print(np.dot(v, W))
print(np.dot(v, W).shape)
```

```python
# This does not. Why?
print(np.dot(W, v))
```

```python
# We can fix the above issue by transposing W.
print(np.dot(W.T, v))
print(np.dot(W.T, v).shape)
```

### Broadcasting

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.

**General Broadcasting Rules**

When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing (i.e. rightmost) dimensions and works its way left. Two dimensions are compatible when:
- they are equal, or
- one of them is 1 (in which case, elements on the axis are repeated along the dimension)

More details: https://numpy.org/doc/stable/user/basics.broadcasting.html

```python
x = np.random.random((3, 4))

y = np.random.random((3, 1))
z = np.random.random((1, 4))

# In this example, y and z are broadcasted to match the shape of x.
# y is broadcasted along dim 1.
s = x + y
# z is broadcasted along dim 0.
p = x * z
```

```python
print(x.shape)
print()
print(y.shape)
print(s.shape)
```

```python
print(x.shape)
print()
print(s.shape)
print(p.shape)
```

```python
a = np.zeros((3, 3))
b = np.array([[1, 2, 3]])
print(a)
print()
print(a+b)
```

### Efficient NumPy Code

When working with numpy arrays, avoid explicit for-loops over indices/axes at all costs. For-loops will dramatically slow down your code (~10-100x).

We can time code using the %%timeit magic. Let's compare using explicit for-loop vs. using numpy operations.

```python
%%timeit
x = np.random.rand(1000, 1000)
for i in range(100, 1000):
    for j in range(x.shape[1]):
        x[i, j] += 5
```

```python
%%timeit
x = np.random.rand(1000, 1000)
x[np.arange(100,1000), :] += 5
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def extract_links(url):
    """
    Verilen URL'den bağlantıları çıkartan fonksiyon.

    Args:
    - url (str): Bağlantıları çıkartılacak URL.

    Returns:
    - set: Çıkartılan bağlantıları içeren küme.
    """
    # URL'yi kontrol etme
    if not urlparse(url).scheme or not urlparse(url).netloc:
        url = "http://" + url

    # HTTP GET isteği gönderme
    try:
        response = requests.get(url)
    except requests.exceptions.RequestException as e:
        print(f"Hata: {e}")
        return set()

    # Sayfa içeriğini BeautifulSoup kullanarak çözümleme
    soup = BeautifulSoup(response.text, 'html.parser')

    # Tüm bağlantıları toplama
    links = set()
    for anchor in soup.find_all('a', href=True):
        link = urljoin(url, anchor['href'])
        links.add(link)

    return links

def crawl_website(start_url, output_file="output_links.txt"):
    """
    Verilen başlangıç URL'sinden başlayarak web sitesini taramak için fonksiyon.

    Args:
    - start_url (str): Taramaya başlanacak başlangıç URL'si.
    - output_file (str): Bağlantıların kaydedileceği dosya adı.
    """
    visited_links = set()
    queue = [start_url]

    while queue:
        current_url = queue.pop(0)

        # Eğer ziyaret edilmişse atla
        if current_url in visited_links:
            continue

        print(f"Taranan URL: {current_url}")

        # URL'den bağlantıları çıkartma
        links = extract_links(current_url)

        # Ziyaret edilen bağlantıları listeye ekleme
        visited_links.add(current_url)

        # Yeni bağlantıları kuyruğa ekleme
        queue.extend(links - visited_links)

        # Bağlantıları dosyaya yazma
        with open(output_file, 'a') as file:
            file.write(f"Taranan URL: {current_url}\n")
            for link in links:
                file.write(f"{link}\n")

    print("Tarama tamamlandı. Bağlantılar dosyaya yazıldı:", output_file)
```

```python
start_url = "http://www.nytimes.com"  # Başlangıç URL'sini değiştirebilirsiniz
output_file = "output_links.txt"   # Çıkartılan bağlantıların kaydedileceği dosya
crawl_website(start_url, output_file)
```

## Nmap tarama
Nmap (Network Mapper), ağdaki bilgisayarları ve hizmetleri keşfetmek için kullanılan bir güvenlik tarayıcısıdır.

```python
import nmap

def nmap_scan_by_hostname(hostname):
    nm = nmap.PortScanner()

    # Belirtilen hostname üzerinde TCP taraması yapma
    nm.scan(hostname, arguments='-p 1-65000')

    # Sonuçları ekrana yazdırma
    for host in nm.all_hosts():
        print(f"Host: {host}")
        print("Port\tState\tService")
        for proto in nm[host].all_protocols():
            lport = nm[host][proto].keys()
            for port in sorted(lport):
                state = nm[host][proto][port]['state']
                service = nm[host][proto][port]['name']
                print(f"{port}\t{state}\t{service}")
```

```python
target_ip = "http://www.nytimes.com"  # Hedef IP adresini değiştirebilirsiniz
nmap_scan_by_hostname(target_ip)
```
