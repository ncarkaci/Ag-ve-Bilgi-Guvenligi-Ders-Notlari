```python
def encode_text_with_spaces(input_text, secret_message, verbose=True):
    """
    Gizli bir mesajı, bir metin içine boşluklar kullanarak kodlar.
    '1' iki boşluk, '0' ise tek boşluk ile temsil edilir.

    Parametreler:
    input_text (str): Gizli mesajın kodlanacağı metin.
    secret_message (str): Kodlanacak gizli mesaj.

    Döndürür:
    str: Kodlanmış metin.
    """
    # Gizli mesajı ASCII değerlerine dönüştür
    ascii_values = [ord(char) for char in secret_message]
    if verbose:
      print("Gizli mesajın ascii değeri:", ascii_values)

    # ASCII değerlerini ikili (binary) formata çevir
    binary_values = [format(ascii_value, '08b') for ascii_value in ascii_values]

    # İkili değerlerin listesini düzleştir
    binary_string = ''.join(binary_values)

    if verbose:
      print("Gizli mesajın binary değeri:", binary_string)

    # İkili stringi giriş metnine kodla
    encoded_text = ''
    binary_index = 0

    for char in input_text:
        if char == ' ' and binary_index < len(binary_string):
            # '1' için çift boşluk, '0' için tek boşluk ekle
            encoded_text += '  ' if binary_string[binary_index] == '1' else ' '
            binary_index += 1
        else:
            encoded_text += char

    # Giriş metninin kalan kısmını ekle
    if binary_index < len(input_text):
        encoded_text += input_text[binary_index:]

    return encoded_text
```

```python
input_text = "Her insan, başka insanları anlama kabiliyetine sahip değildir. Bunu ancak, öteki insanların gönül dünyalarında yaşayabilenler başarır. Bu insanlar, başkalarının sevinçlerine, üzüntülerine ortak olabilir, onların acılarını hissedebilir, onlarla birlikte ağlayıp, onlarla birlikte gülebilirler. Fakat herkes bu duyarlılığa sahip değildir. Bazıları vardır ki, sadece kendi benlikleri içinde yaşarlar ve başkalarının ne hissettiğini anlamaktan acizdirler."
secret_message = "Gizli plan"

encoded_text_example = encode_text_with_spaces(input_text, secret_message)
print(encoded_text_example)
```
