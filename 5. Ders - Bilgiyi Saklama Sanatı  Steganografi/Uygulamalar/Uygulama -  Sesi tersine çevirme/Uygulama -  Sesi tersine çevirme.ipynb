```python
import pydub
import os

def reverse_audio_mp3(input_file):
    # Orijinal dosya adını ve uzantısını ayırma
    file_name, file_extension = os.path.splitext(input_file)
    output_file = f"{file_name}_reversed{file_extension}"

    # MP3 dosyasını yükleme
    audio = pydub.AudioSegment.from_file(input_file)

    # Ses dosyasını tersine çevirme
    reversed_audio = audio.reverse()

    # Ters çevrilmiş ses dosyasını kaydetme
    reversed_audio.export(output_file, format=file_extension.strip('.'))

    return output_file
```

```python
input_audio_mp3 = './ters_ses_dosyası.mp3'
output_audio_mp3 = reverse_audio_mp3(input_audio_mp3)
```
