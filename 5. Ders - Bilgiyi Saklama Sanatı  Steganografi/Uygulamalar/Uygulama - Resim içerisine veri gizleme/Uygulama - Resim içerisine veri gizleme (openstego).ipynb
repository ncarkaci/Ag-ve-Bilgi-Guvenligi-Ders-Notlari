### Uygulama

To embed `secret.txt` into `wallpaper.png` and generate the output into `stego.png`:



```python
! java -jar ./openstego-0.8.6/lib/openstego.jar embed -a randomlsb -mf secret.txt -cf cover.png -sf stego.png

# or
#! java -jar <path>\openstego.jar --embed --algorithm=randomlsb --messagefile=secret.txt --coverfile=wallpaper.png --stegofile=test.png
```

To extract embedded data from `stego.png` to `/data` directory:

```python
! mkdir data
! java -jar ./openstego-0.8.6/lib/openstego.jar extract -a randomlsb -sf stego.png -xd ./data

# or

# java -jar ./openstego-0.8.6/lib/openstego.jar extract --algorithm=randomlsb --stegofile=stego.png --extractdir=/data
```

To generate a signature file:

```python
! java -jar ./openstego-0.8.6/lib/openstego.jar --gensig --algorithm=dwtdugad --sigfile=my.sig
```

To embed signature into `cover.png` and generate the output into `
signed_cover.png`:

```python
! java -jar ./openstego-0.8.6/lib/openstego.jar --embedmark --algorithm=dwtdugad --sigfile=my.sig --coverfile=cover.png --stegofile=signed_cover.png
```

To check for watermark in `signed_cover.png` using `my.sig` signature file:

```python
! java -jar ./openstego-0.8.6/lib/openstego.jar checkmark --algorithm=dwtdugad --sigfile=my.sig --stegofile=signed_cover.png
```
