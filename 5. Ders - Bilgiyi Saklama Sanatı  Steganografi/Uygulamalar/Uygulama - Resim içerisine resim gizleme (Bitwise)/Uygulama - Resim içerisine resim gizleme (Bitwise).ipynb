```python
from PIL import Image, ImageEnhance
import numpy as np

# 1. Ağaç resmini yükleyin
image_path = "./images/tree.png"
original_image = Image.open(image_path)
```

```python
# 2. Resmin her baytına 3 sayısı ile "Logical AND" işlemi yapın
image_array = np.array(original_image)
and_result = np.bitwise_and(image_array, 3)
```

```python
# 3. Siyah bir resmi oluşturun
black_image = Image.fromarray(and_result.astype(np.uint8))

# 4. Siyah resmi kaydetmek için
black_image_path = "./images/black_image.jpg"
black_image.save(black_image_path)
```

```python
from IPython.display import Image
Image('./images/black_image.jpg')
```

```python
# 5. Resmin parlaklığını 85 katına çıkarın
brightness_enhancer = ImageEnhance.Brightness(black_image)
bright_image = brightness_enhancer.enhance(85)

# 6. Parlak resmi kaydetmek için
bright_image_path = "./images/cat.jpg"
bright_image.save(bright_image_path)
```
